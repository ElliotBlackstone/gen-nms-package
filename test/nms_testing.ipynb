```python
import torch
import gen_nms

boxes = torch.tensor([
    [0.0, 0.0, 10.0, 10.0],
    [1.0, 1.0, 11.0, 11.0],
    [50.0, 50.0, 60.0, 60.0],
], dtype=torch.float32)

scores = torch.tensor([0.91, 0.82, 0.73], dtype=torch.float32)
labels = torch.tensor([0, 0, 1], dtype=torch.int64)

for name in ["iou_nms", "giou_nms", "diou_nms", "ciou_nms"]:
    cpu_out = getattr(gen_nms, name)(boxes, scores, 0.5)
    gpu_out = getattr(gen_nms, name)(boxes.cuda(), scores.cuda(), 0.5).cpu()
    print(name, "cpu:", cpu_out, "gpu:", gpu_out, "match:", torch.equal(cpu_out, gpu_out))

for name in ["batched_iou_nms", "batched_giou_nms", "batched_diou_nms", "batched_ciou_nms"]:
    cpu_out = getattr(gen_nms, name)(boxes, scores, labels, 0.5)
    gpu_out = getattr(gen_nms, name)(boxes.cuda(), scores.cuda(), labels.cuda(), 0.5).cpu()
    print(name, "cpu:", cpu_out, "gpu:", gpu_out, "match:", torch.equal(cpu_out, gpu_out))
```

```python
import torch
import gen_nms

def sample_inputs(device):
    boxes = torch.tensor(
        [
            [0.0, 0.0, 10.0, 10.0],
            [1.0, 1.0, 11.0, 11.0],
            [50.0, 50.0, 60.0, 60.0],
            [52.0, 52.0, 61.0, 61.0],
        ],
        dtype=torch.float32,
        device=device,
    )
    scores = torch.tensor([0.9, 0.8, 0.7, 0.6], dtype=torch.float32, device=device)
    threshold = 0.5
    return (boxes, scores, threshold)

ops = [
    torch.ops.gen_nms.iou_nms,
    torch.ops.gen_nms.giou_nms,
    torch.ops.gen_nms.diou_nms,
    torch.ops.gen_nms.ciou_nms,
]

for op in ops:
    torch.library.opcheck(op, sample_inputs("cpu"))
    print(f"{op} CPU opcheck passed")

if torch.cuda.is_available():
    for op in ops:
        torch.library.opcheck(op, sample_inputs("cuda"))
        print(f"{op} CUDA opcheck passed")
```

```python
import torch
from torch._subclasses.fake_tensor import FakeTensorMode
import gen_nms

boxes = torch.randn(8, 4, dtype=torch.float32)
boxes[:, 2:] += boxes[:, :2].abs() + 1.0  # make x2>x1, y2>y1
scores = torch.rand(8, dtype=torch.float32)

mode = FakeTensorMode()
converter = mode.fake_tensor_converter

fake_boxes = converter.from_real_tensor(mode, boxes)
fake_scores = converter.from_real_tensor(mode, scores)

with mode:
    out = torch.ops.gen_nms.ciou_nms(fake_boxes, fake_scores, 0.5)

print(type(out))
print("device:", out.device)
print("dtype:", out.dtype)
print("ndim:", out.ndim)
print("shape:", out.shape)
```

```python
import torch
import gen_nms

def f(boxes, scores):
    return torch.ops.gen_nms.ciou_nms(boxes, scores, 0.5)

boxes = torch.tensor(
    [[0., 0., 10., 10.],
     [1., 1., 11., 11.],
     [50., 50., 60., 60.]],
    dtype=torch.float32,
)
scores = torch.tensor([0.9, 0.8, 0.7], dtype=torch.float32)

compiled_f = torch.compile(f, fullgraph=True)
out_eager = f(boxes, scores)
out_compiled = compiled_f(boxes, scores)

print(torch.equal(out_eager, out_compiled))
```

```python
import torch
import gen_nms

class M(torch.nn.Module):
    def forward(self, boxes, scores):
        return torch.ops.gen_nms.ciou_nms(boxes, scores, 0.5)

m = M()
boxes = torch.tensor(
    [[0., 0., 10., 10.],
     [1., 1., 11., 11.],
     [50., 50., 60., 60.]],
    dtype=torch.float32,
)
scores = torch.tensor([0.9, 0.8, 0.7], dtype=torch.float32)

ep = torch.export.export(m, (boxes, scores))
print(ep)
```
